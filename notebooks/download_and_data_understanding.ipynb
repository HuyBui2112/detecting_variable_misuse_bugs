```python
from pathlib import Path
import os

REPO_URL = "https://github.com/HuyBui2112/detecting_variable_misuse_bugs.git"
WORKSPACE = Path("/content/drive/MyDrive/is353_variable_misuse")
REPO_DIR = WORKSPACE / "detecting_variable_misuse_bugs"
DATASET_ROOT = WORKSPACE / "dataset"

WORKSPACE.mkdir(parents=True, exist_ok=True)
print("WORKSPACE:", WORKSPACE)
print("REPO_DIR:", REPO_DIR)
print("DATASET_ROOT:", DATASET_ROOT)
```

```python
import subprocess
import os

if not REPO_DIR.exists():
    assert REPO_URL != "PASTE_YOUR_GIT_REPO_URL_HERE", "Bạn cần điền REPO_URL trước."
    subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=True)
else:
    os.chdir(REPO_DIR)
    subprocess.run(["git", "pull"], check=True)

os.chdir(REPO_DIR)
os.environ["PYTHONPATH"] = str(REPO_DIR / "src")
print("Current dir:", os.getcwd())
```
